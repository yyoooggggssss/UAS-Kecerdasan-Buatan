<font color = "blue"> Method 1.2</font>


<font color = "green"> Using Plillow (PIL) with matplolib Module</font>

```python
from PIL import Image

import matplotlib.pyplot as plt

img = Image.open('./saitama.jpg')

plt.imshow(img)
```

<font color = "blue"> Method 2</font>

<font color = "green"> Using Matplotlib Module</font>

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('./saitama.jpg')

plt.imshow(img)
```
